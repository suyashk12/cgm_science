```python
# Define a function to load in the spectrum of a given galaxy
def load_spec(rootdir, gal_id):
    
    # Construct a wavelength array
    wav_0 = data_hdu.header['CRVAL3']
    del_wav = data_hdu.header['CD3_3']
    wav_arr = np.arange(wav_0, wav_0 + (spec_cube.shape[0]-1)*del_wav, del_wav)
    
    # Isolate coordinates from segmentation map
    seg_map_gal_y, seg_map_gal_x = np.where(seg_map==gal_id)

    # Get bolometric fluxes for each pixel in the galaxy
    flux_gal = np.zeros(len(seg_map_gal_x))
    
    # Also isolate the spectra for each pixel
    spec_stack = np.zeros((len(seg_map_gal_x), len(wav_arr)))
    
    for j in range(len(seg_map_gal_x)):
        
        # Get the x and y coordinates
        x = seg_map_gal_x[j]
        y = seg_map_gal_y[j]

        # Get the bolometric flux for this coordinate
        flux_gal[j] = np.nansum(spec_cube[:,y,x])
        
        # Also isolate the spectrum for the pixel
        spec_stack[j,:] = spec_cube[:,y,x]
        
    # These are some bolometric flux cuts to include/ exclude pixels when stacking the spectra
    min_flux_gal = np.nanpercentile(flux_gal, 75)
    max_flux_gal = np.nanpercentile(flux_gal, 100)
    
    idx_bright = (flux_gal>=min_flux_gal) & (flux_gal<=max_flux_gal)
    
    # Undo normalization before saving the spectrum
    f_lam = np.nansum(spec_stack[idx_bright], axis=0)*1e-20
    
    return wav_arr, f_lam
```

From CUBS I, we have the following bandpasses -

<blockquote>4800–5800 Å (pseudo g-band), 6000–7000 Å (pseudo r-band), and 7500–8500 Å (pseudo i-band)</blockquote>

But in CUBS VI, we modify the bandpass to be -

<blockquote>4800-5800 Å (pseudo g-band), 6000-7500 Å (pseudo r-band), and 7500-9000 Å (pseudo i-band)</blockquote>

Let's try performing this calculation for just one galaxy first

```python
# Define this for automation later
bandpasses_lam = {'pseudo_g': [4800,5800], 'pseudo_r': [6000,7500], 'pseudo_i': [7500,9000]}
```

```python
# Load in the spectrum for a single galaxy
wav, f_lam = load_spec(rootdir, 73)
```

```python
# Plot the spectrum to get a sense
fig, ax = plt.subplots(1, figsize=(15,7))

ax.step(wav, f_lam, where='mid', color='black')

# Also highlight the bandpasses
ax.axvspan(bandpasses_lam['pseudo_g'][0], bandpasses_lam['pseudo_g'][1], color='greenyellow')
ax.axvspan(bandpasses_lam['pseudo_r'][0], bandpasses_lam['pseudo_r'][1], color='pink')
ax.axvspan(bandpasses_lam['pseudo_i'][0], bandpasses_lam['pseudo_i'][1], color='lightcyan')

ax.set_xlabel('Wavelength (Å)')
ax.set_ylabel('Flux (ergs/cm^2/s/Å)')
```

We use the following conversion from $f_{\lambda}$ to $f_{\nu}$ -

$$f_{\nu} = \frac{\lambda^2}{c} f_{\lambda} \\
\Rightarrow \frac{f_{\nu}}{Jy} = 3.34 \times 10^4 \left(\frac{\lambda}{\text{Å}}\right)^2 \frac{f_{\lambda}}{\text{ergs cm}^{-2} \text{s}^{-1} \text{Å}^{-1}}$$

```python
# Let us remake the spectrum but in frequency space

c = 3e+8 # Light speed in m/s

nu = c/(wav*1e-10) # in Hz
f_nu = wav**2*f_lam*(3.34e+4) # Convert from ergs/s/cm^2/Å then convert to Janskys
```

```python
# Also convert bandpasses to frequency space

bandpasses_nu = {}

for band in list(bandpasses_lam.keys()):
        
    bandpasses_nu[band] = [c/(bandpasses_lam[band][1]*1e-10), c/(bandpasses_lam[band][0]*1e-10)]
```

```python
# Plot the spectrum to get a sense
fig, ax = plt.subplots(1, figsize=(15,7))

ax.step(nu, f_nu, where='mid', color='black')

# Also highlight the bandpasses
ax.axvspan(bandpasses_nu['pseudo_g'][0], bandpasses_nu['pseudo_g'][1], color='greenyellow')
ax.axvspan(bandpasses_nu['pseudo_r'][0], bandpasses_nu['pseudo_r'][1], color='pink')
ax.axvspan(bandpasses_nu['pseudo_i'][0], bandpasses_nu['pseudo_i'][1], color='lightcyan')

ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('Flux (Janskys)')
```

We define the AB magnitude for a monochromatic source as -

$$m_{AB} = -2.5 \log\left(\frac{f_{\nu}}{Jy} \right) + 8.9$$

Where $f_{\nu}$ is expressed in Janskys. Since we are dealing with a bandpass instead of a discrete frequency, and we are dealing with a CCD (photon-counting device), we'll instead use

$$m_{AB} = -2.5 \log\left( \frac{\int f_\nu (h \nu)^{-1} e(\nu) \mathrm{d} \nu}{\int (3631 \ \text{Jy}) (h \nu)^{-1} e(\nu) \mathrm{d} \nu} \right)$$

Where $e(\nu)$ is the equal energy bandpass function, and the division by $h\nu$ achieves converting the spectral flux density to a photon count rate.

```python
AB_mags = {}

for band in list(bandpasses_nu.keys()):
    
    mask = ((nu>bandpasses_nu[band][0])&(nu<bandpasses_nu[band][1]))
        
    nu_mask = nu[mask]
        
    f_nu_mask = f_nu[mask]
    
    mag = -2.5*np.log10(np.trapz(y=f_nu_mask/nu_mask, x=nu_mask)/np.trapz(y=3631/nu_mask, x=nu_mask))    
    #mag = -2.5*np.log10(np.trapz(y=f_nu_mask, x=nu_mask)/(nu_mask[-1]-nu_mask[0])) + 8.9
    
    
    AB_mags[band] = mag
```
