```python
query_ra = Angle("15h24m26.3325s").rad
query_dec = Angle("+9d58m21.062s").rad
```

```python
def calc_phi(alpha_1, delta_1, alpha_2, delta_2):
    
    cos_phi = np.dot([np.cos(delta_1)*np.cos(alpha_1), np.cos(delta_1)*np.sin(alpha_1), np.sin(delta_1)],
                      [np.cos(delta_2)*np.cos(alpha_2), np.cos(delta_2)*np.sin(alpha_2), np.sin(delta_2)])
    
    phi = np.arccos(cos_phi)
    
    return phi
```

```python
phi_arr = []

for i in range(len(ldss_photo_df)):
    
    ra = Angle(ldss_photo_df.loc[i]['RA'], units.deg).rad
    dec = Angle(ldss_photo_df.loc[i]['Dec'], units.deg).rad
    
    phi_arr.append((calc_phi(ra, dec, query_ra, query_dec)*units.rad).to(units.arcsecond).value)
    
phi_arr = np.array(phi_arr)
```
