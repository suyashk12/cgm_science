# z=1.09

For this absorber, we detect significant absorption for \ion{H}{I}, \ion{C}{III}, \ion{C}{IV}, \ion{N}{IV}, \ion{O}{III}, \ion{O}{IV}, \ion{O}{V}, and \ion{O}{VI}. The HIRES \ion{C}{IV} absorption reveals a single component structure at $\varv_c \approx 0$ km/s. \ion{O}{IV} 553/554/608/787 are well-aligned with the \ion{C}{IV} absorption, providing confidence in the velocity centroid. The \ion{C}{III} absorption in COS NUV is higher S/N than the STIS coverage but requires a velocity offset of $\delta \varv\!=\!-12.5 \ \mathrm{km/s}$ to align it with other ions. The \ion{C}{III} transition is also saturated, leading to degeneracy in the inferred column density and linewidth. For this reason, we tie the \ion{C}{III} linewidth to the well-determined \ion{C}{IV} linewidth to derive a column density constraint on the former. During simultaneous fitting of all ions, we require the \ion{C}{III} column density to be lower than the corresponding \ion{H}{I} column density during fitting, as \ion{C}{III} larger than the associated \ion{H}{I} content would require an unreasonably high carbon abunance. \ion{N}{IV} 765 requires a velocity offset of $\delta \varv \!=\! -5.3 \ \mathrm{km/s}$. \ion{O}{III} 702 requires a velocity offset $\delta \varv\!=\!2.1 \ \mathrm{km/s}$ and \ion{O}{III} 832 requires a velocity offset of $\delta \varv = -4.3 \ \mathrm{km/s}$ to align them with each other and other ions. As for high ions, \ion{O}{V} 629 requires a velocity offset of $\delta \varv \!=\! 6.7 \ \mathrm{km/s}$; applying this shift is justified given its consistent linewidth with \ion{O}{IV} and \ion{O}{III}. Note that \ion{O}{VI} 1031 was contaminated with Ly$\gamma$ at $z = 1.23470$, but we correct this contamination; the \ion{O}{VI} doublet is then found to be at $\varv_c \! \approx \! 0 \ \mathrm{km/s}$.

For \ion{H}{I} 1215/1025, a single component aligned with metal absorption cannot explain the complete absorption profile. We thus find it necessary to invoke a blueward metal-free \ion{H}{I} component (c1), imposing it to be weaker than the component aligned with metals, i.e. $N_\mathrm{HI, c1} < N_\mathrm{HI, c2}$, where c2 is the component bearing metal absorption.

With the above restrictions, we perform Voigt profile fitting for the detected ions, the results for which are reported in Table \ref{tab:z_109} and the top panel of Figure \ref{fig:z_109}. Given similar linewidths across a wide range of ionization potentials, we adopt $b\!=\!20 \ \mathrm{km/s}$ for all non-detected ions.

For ionization modeling of c2, we first note that consistent linewidths for \ion{O}{V} and \ion{O}{IV} suggest that they may arise in the same phase. Using the ratio \ion{O}{V}/\ion{O}{IV}, a single gas phase of density $\log(n_\mathrm{H,c2L}/\mathrm{cm}^{-3}) \!\approx\! -3.1$ underpredicts the ratio \ion{O}{III}/\ion{O}{IV}, indicating that the remaining \ion{O}{III} arise in a denser phase. The low-density phase also underpredicts \ion{O}{VI}, which may suggest its origin in a more ionized phase. However, the narrow linewidth of \ion{O}{VI} and the low S/N could imply that the identified absorption is contamination from another redshift. For this reason, we only incorporate the \ion{O}{VI} measurement as an upper limit for ionization modeling. The high-density phase containing \ion{O}{III} can be constrained by treating \ion{O}{IV}/\ion{O}{III} as an upper limit, yielding $\log(n_\mathrm{c2H}/\mathrm{cm}^{-3})\!>\!-3.2$, while the \ion{He}{I}/\ion{H}{I} upper limit dictates $\log(n_\mathrm{c2H}/\mathrm{cm}^{-3})\!<\!-3.0$. The narrower linewidth of \ion{C}{IV} and \ion{C}{III} suggest that they should arise in a cooler, denser; in fact, the gas density recovered from the \ion{C}{IV}/\ion{C}{III} ration, $\log(n_\mathrm{c2H}/\mathrm{cm}^{-3})\!\approx\!-3.1$ is consistent with the limits imposed from \ion{O}{IV}/\ion{O}{III} and \ion{He}{I}/\ion{H}{I} upper limits.

We perform a two-phase PIE fit, recovering the high-density phase at $\log(n_\mathrm{H,c2H}/\mathrm{cm}^{-3}) \! \approx \! -3.1$ and the low-density phase at $\log(n_\mathrm{H,c2L}/\mathrm{cm}^{-3}) \! \approx \! -3.8$. \ion{O}{III} is almost entirely attributed to the high-density phase, constraining its metallicity at $\mathrm{[\alpha/H]}_\mathrm{c2H} \approx 0.2$. Since \ion{C}{III} and \ion{C}{IV} are attributed to the high-density phase, we recover its relative carbon abundance $\mathrm{[C/\alpha]}_\mathrm{c2H} \approx 0.3$. The cloud size of the high-density phase is constrained to be $l_\mathrm{c2H} \approx 0.2 \ \mathrm{kpc}$. For the low-density phase, we obtain a range of metallicities by imposing limits on its cloud size, which yields $-1.1 \!<\! [\mathrm{\alpha/H}]_\mathrm{c2L} \!<\! 1.0$. 

However, there are a few inconsistencies of the two-phase PIE model. The best-fit model overpredicts upper limits for \ion{He}{I}, \ion{Si}{III}, \ion{S}{IV} and \ion{S}{V}. While $\mathrm{[Si/O]}$ may be affected by dust depletion, alleviating the tension between the model prediction and the upper limit on \ion{Si}{III}, we should expect $\mathrm{[S/O]}=0$, meaning that the \ion{S}{IV} and \ion{S}{V} upper limits should not be violated. Another issue is the high metallicity of the low-density phase; the posterior for the metallicity peaks around $\mathrm{[\alpha/H]}_\mathrm{c2L} \!\approx\! 0.4$, and cuts off at $\mathrm{[\alpha/H]}_\mathrm{c2L} \!=\! 1.0$ because we do not run PIE models with higher metallicities, meaning that even higher metallicities are possible. This issue arises in part because of the modest value for the total measured $N_\mathrm{HI,c2}$, which subsequently puts a stringent upper limit on the associated \ion{H}{I} column density for the low-density phase, resulting in metallicities $\mathrm{[\alpha/H]}_\mathrm{c2L} \!>\! -1.1$. A direct consequence of high metallicity is a short gas cooling time. For PIE, gas cooling times are typically larger than the ionic recombination time, as evaluated for our sample. However, for the low-density phase of c2, using the inferred density and metallicity posteriors yields $\log(t_\mathrm{cool}/t_\mathrm{rec}) = 0.1 \pm 0.3$, making the cooling time consistent with the recombination time. This is inconsistent with the ionization equilibrium assumption of PIE.

For this reason, we consider modeling the low-density phase using TDP. In TDP, ion fractions are dependent on gas density, metallicity, and temperature. Therefore, the low-density phase, in addition to having an associated \ion{H}{I} column density, gas density, metallicity, and relative abundances, also has an associated temperature. By using absorption linewidths for \ion{O}{IV} and \ion{O}{V}, we can place $3\sigma$ limits on the gas temperature for the low-density phase, $4.1\!<\!\log(T_\mathrm{c2L}/\mathrm{K})\!<\!5.8$, which we use to inform our temperature prior for the low-density TDP phase. Using the diagnostic plot in Figure \ref{fig:diagnostic}, we find that unlike PIE, a single TDP phase with $\log(n_\mathrm{c2L}/\mathrm{cm}^{-3}) \!\approx\! -3.5$, $\mathrm{[\alpha/\mathrm{H}]}_\mathrm{c2L} \!\approx\! 0.0$, and $\log(T_\mathrm{c2L}/\mathrm{K}) \!\approx\! 4.9$ can simultaneously explain the measured ratios for \ion{O}{V}/\ion{O}{IV} and \ion{O}{IV}/\ion{O}{III}. This prediction is also consistent with the similar linewidths for \ion{O}{III}, \ion{O}{IV}, and \ion{O}{V}. However, given narrower linewidths for \ion{C}{IV} and \ion{C}{III}, we still attribute them to the high-density phase in PIE. Note that because the \ion{O}{III} is now attributed to the low-density TDP phase, the associated metallicity $\mathrm{[\alpha/H]}_\mathrm{c2H}$ of the high-density phase will be lowered. While this can address the tension with \ion{Si}{III}, \ion{S}{IV}, and \ion{S}{V} upper limits, the relative carbon abundance $[\mathrm{C/\alpha}]_\mathrm{c2H}$ will now be higher than the two-phase TDP model.

We perform a simultaneous two-phase fit for this absorption system, with the high-density phase in PIE and the low-density phase in TDP. Similar to the two-phase PIE fit, we attribute the measured $N_\mathrm{HI,c2}$ to the high-density phase. The gas density for the high-density phase in PIE is still dictated by the ratio \ion{C}{IV}/\ion{C}{III} and is estimated to be $\log(n_\mathrm{H,c2H}/\mathrm{cm}^{-3}) \approx -3.1$. However, the metallicity estimate for the high-density phase is made uncertain because most of the measured \ion{O}{III} can be attributed to the low-density phase, unlike the two-phase PIE fit. Decreasing $\mathrm{[\alpha/H]}_\mathrm{c2H}$ will however require a correspondingly high $[\mathrm{C/\alpha}]_\mathrm{c2H}$ to reproduce the measured \ion{C}{III} and \ion{C}{IV} column densities. Based on measurements of $[\mathrm{C/\alpha}]$ in the $z \!\lesssim\! 1$ CGM (\citealt{Zahedy21}, \citealt{Cooper21}), we restrict $[\mathrm{C/\alpha}]_\mathrm{c2H} \!\le\! 1.5$ during the simultaneous fitting, which imposes a corresponding lower limit on $\mathrm{[\alpha/H]}_\mathrm{c2H}$. Furthermore, the \ion{Si}{III} upper limit provides an upper limit on $\mathrm{[\alpha/H]}_\mathrm{c2H}$, given the well-determined gas density for this phase. Through simultaneous fitting, we recover a metallicity of $\mathrm{[\alpha/H]}_\mathrm{c2H} \approx -0.3$ and carbon relative abundance of $\mathrm{[C/\alpha]}_\mathrm{c2H} \approx 0.7$, both with uncertainties of $\approx\!0.3 \ \mathrm{dex}$ in both quantities. This still leaves most of \ion{O}{III}, and all of \ion{O}{V} and \ion{O}{VI} to be produced by the low-density TDP phase. 

The gas density of c2L in TDP, like PIE, is still well constrained by the measured column density ratios. This may be understood in terms of the diagnostic plot presented in Figure \ref{fig:diagnostic}. Changing gas density by $\approx \! 0.2 \ \mathrm{dex}$ also modifies the \ion{O}{V}/\ion{O}{IV} ratio depicted on the $y$-axis of the diagnostic by about $\approx \! 0.2 \ \mathrm{dex}$, irrespective of the gas density or metallicity. Therefore, the well measured \ion{O}{V}/\ion{O}{IV} ratio constrains the gas density at $\log(n_\mathrm{H,c2L}/\mathrm{cm}^{-3}) \! \approx \! -3.6$, consistent with the PIE prediction. Once gas density is constrained using the \ion{O}{V}/\ion{O}{IV} ratio, the gas temperature can be constrained using the \ion{O}{IV}/\ion{O}{III} ratio on the $x$-axis of the diagnostic. Given that TDP ion fractions also depend on gas metallicity, there is some degeneracy between the gas temperature and metallicity. To resolve this degeneracy, we fit instead the measured \ion{O}{III}, \ion{O}{IV}, and \ion{O}{V} column densities with a TDP phase while imposing restrictions on its cloud size, similar to the two-phase PIE model. Namely, we require the low-density phase under TDP to be larger than its sonic length and the compact high-density phase, while being smaller than its Jeans length, as permitted within the error budget of the measured $N_\mathrm{HI,c2}$. The subsequent limits of gas metallicities are recovered at $-1.3 \!<\! [\mathrm{\alpha/H}]_\mathrm{c2L} \!<\!0.7$, while the gas temperature is constrained to be $\log(T_\mathrm{c2L,TDP}/\mathrm{K})\!\approx\!4.9$. The gas metallicities are systematically lower relative to PIE estimate, which can be understood in terms larger permitted gas temperatures. While the PIE temperature was estimated at $\log(T_\mathrm{c2L,PIE}/\mathrm{K})\!\approx\!4.0$, higher gas temperatures in TDP can enhance ion fractions, implying that measured column densities under TDP can be explained using lower metallicities, following Equation \ref{eqn:col_dens}. The inferred ratio of gas cooling time to recombination time for the low-density phase  under TDP, $\log(t_\mathrm{cool}/t_\mathrm{rec}) = 0.5_{-0.4}^{+0.3}$, implies that the gas can still cool on timescales comparable to ionic recombination, but such scenarios are self-consistently handled under TDP, unlike PIE. 

In conclusion, assuming the low-density phase for c2 to be in PIE results in large gas metallicities which manifest themselves as short cooling times, a regime in which the ionization equilibrium assumption for PIE can be challenged. Upon using TDP for the low-density phase instead, high gas metallicities and short cooling times can be modeled self-consistently. However, the simultaneous allocation of \ion{O}{III}-\ion{O}{V} to the low-density TDP phase lowers metallicity for the high-density phase, requiring a high relative carbon abundance, $[\mathrm{C/\alpha}]_\mathrm{c2H} = 0.7 \pm 0.3$ to reproduce the measured \ion{C}{III} and \ion{C}{IV} column densities. Such a high carbon relative abundance is not unreasonable given the near solar gas metallicity of $[\mathrm{\alpha/H}]_\mathrm{c2H} \approx -0.3_{-0.3}^{+0.2}$, and may indicate a secondary enrichment mechanism like stellar/AGB winds. We also note that the high relative carbon abundance does not affect the PIE gas cooling curve relevant for c2H, but we defer a more detailed discussion of relative abundances on gas cooling to Kumar et al. (in prep). 

# z=1.16

For this absorber, we identify significant absorption for \ion{H}{I}, \ion{C}{III}, \ion{C}{IV}, \ion{O}{III}, \ion{O}{IV}, \ion{O}{V}, \ion{O}{VI}, and \ion{Si}{III}. The HIRES \ion{C}{IV} shows a two-component structure, with a a broad component (c1) at $\varv_c \!\approx\! -25 \ \mathrm{km/s}$ and a narrow component (c1) at $\varv_c \!\approx\! 0 \ \mathrm{km/s}$ and . This two-component structure is also revealed in \ion{O}{IV}, although the restriction $b_{\mathrm{OIV,c1}}\!>\!b_{\mathrm{OIV,c2}}$ needs to be imposed during fitting. Likewise, \ion{O}{V} also exhibits this two component structure, but $N_{\mathrm{OV,c1}}\!>\!N_{\mathrm{OV,c2}}$ needs to be imposed. \ion{O}{VI} reveals a single broad component consistent in centroid with c1. \ion{O}{III} 702 can be aligned with c2 by applying an offset of $\delta \varv = 2.6 \ \mathrm{km/s}$; \ion{O}{III} 832 is contaminated and cannot be included in the fitting. Both \ion{C}{III} 977 and \ion{Si}{III} 1206 are covered by STIS, but require an offset of $\delta \varv = 6 \ \mathrm{km/s}$ to align them with c2. During simultaneous fitting, we also restrict the \ion{C}{III} column density to be lower than the corresponding \ion{H}{I} column density while also tying its linewidth to the corresponding \ion{C}{IV} component to derive a column density constraint given the saturated transition.

\ion{H}{I} is decomposed into three components, with two components consistent in centroid with c1 and c2, while the third redward component c3 is metal-free. During fitting, we have to restrict $b_\mathrm{HI,c2}\!<\!b_\mathrm{HI,c3}\!<b_\mathrm{HI,c1}$ and $N_\mathrm{HI,c3}\!<\!N_\mathrm{HI,c1}$. We also see detections for \ion{He}{I} 584/537 in COS, but offsets of $\delta \varv = 1 \ \mathrm{km/s}$ and $\delta \varv = 10.3 \ \mathrm{km/s}$ respectively are necessary to align them with c2.

The results of simultaneous Voigt profile fitting are presented in the top panel of Figure \ref{fig:z_116} and Table \ref{tab:z_116}. The metal-free HI component is constrained to be at $\varv_{\mathrm{c,c3}} = 21_{-8}^{+6} \ \mathrm{km/s}$ with $\log(N_\mathrm{HI,c3}/\mathrm{cm}^{-2}) = 14.5_{-0.3}^{+0.2}$ and $b_\mathrm{HI,c3} = 22_{-4}^{+5} \ \mathrm{km/s}$. In the second panel of Figure \ref{fig:z_116} from the top, we overlay the best-fit \ion{H}{I} profile against the higher-order Lyman series covered by COS, finding the modest flux decrement to be consistent with the observed spectrum.

For PIE modeling, we find a single gas phase to be sufficient for modeling the two metal absorption components, although the ionizing conditions are found to be quite different between the two components. For c1, a gas density of $\log(n_\mathrm{H,c1}/\mathrm{cm}^{-3}) \approx -3.2$ set by the \ion{O}{V} to \ion{O}{IV} column density ratio is also consistent with the \ion{O}{III}/\ion{O}{IV}, \ion{Si}{III}/\ion{O}{IV}, and \ion{He}{I}/\ion{H}{I} column density ratios, providing high confidence in the inferred density.
Given the large \ion{H}{I} content, the inferred metallicity $\mathrm{[\alpha/H]}_\mathrm{c1} \approx -1.8$ is low. On the other hand, for c2, the gas density set by \ion{O}{IV} and \ion{O}{V} is consistent with the non-detection of \ion{O}{III} but under-produces \ion{O}{VI}, likely indicating a more ionized gas phase. This is consistent with \ion{O}{VI} having a broader linewidth than other ions, so we do not attempt to model the \ion{O}{VI} absorption. The inferred gas metallicity for the second component $\mathrm{[\alpha/H]}_\mathrm{c2} \approx -1.6$ is roughly consistent with c1. However, the relative carbon abundance $\mathrm{[C/\alpha]}_\mathrm{c2} \approx 0$ is larger than c1, which has $\mathrm{[C/O]}_\mathrm{c1} \approx -0.2$. The inferred cloud sizes are large, with $\log(l_\mathrm{c1}/\mathrm{kpc}) \approx 1.5$ and $\log(l_\mathrm{c2}/\mathrm{kpc}) \approx 1.7$. Both components are found to be dominant in non-thermal broadening, which imposes a temperature upper limit consistent with the PIE inferred temperature. Additionally, the gas cools on timescales longer than ionic recombination for both components. Therefore, we do not pursue TDP modeling for this system. 

# z=1.22

For this absorber, we find significant detection for \ion{H}{I}, \ion{C}{IV}, \ion{O}{IV}, \ion{O}{V}, and \ion{O}{VI}. Both \ion{C}{IV} and \ion{O}{IV} are well-described by a single component (c1) at $\varv_c \approx 0 \ \mathrm{km/s}$. For \ion{O}{V}, a shift of $\delta \varv = -5.2 \ \mathrm{km/s}$ aligns it with $\delta \varv = 4.4 \ \mathrm{km/s}$  \ion{O}{VI} 1031 aligns it with c1; \ion{O}{VI} 1037 is contaminated, so we do not include it during the fitting.

\ion{H}{I} 1215/1025 are decomposed into four components to fully explain the absorption profiles. One of these components is aligned with metal absorption while the other three are metal-free (labeled c2, c3, c4 going redward). Given the limited number of transitions detected for \ion{H}{I}, several restrictions on the prior are necessary to derive significant constraints for each component. We first need to impose $\varv_\mathrm{c,c2}\!<\!\varv_\mathrm{c,c1}\!<\!\varv_\mathrm{c,c3}\!<\!\varv_\mathrm{c,c4}$, $\varv_\mathrm{c2}\!<\!-20 \ \mathrm{km/s}$, and $\varv_\mathrm{c3}\!>\!20 \ \mathrm{km/s}$. Next, we impose $b_\mathrm{HI,c1}\!<\!b_\mathrm{HI,c3}\!<\!b_\mathrm{HI,c2}\!<\!b_\mathrm{HI,c4}$. Finally, we require $N_\mathrm{HI,c1}\!>\!N_\mathrm{HI,c2}$ and $N_\mathrm{HI,c3}\!>\!N_\mathrm{HI,c4}$.

We present the results of simultaneous Voigt profile fitting in the top panel of Figure \ref{fig:z_122} and Table \ref{tab:z_122}. The bluest metal-free \ion{H}{I} components is constrained to be at $\varv_\mathrm{c,c2} = -32 \pm 5 \ \mathrm{km/s}$, with $\log(N_\mathrm{HI,c2}/\mathrm{cm}^{-2}) = 13.89 \pm 0.08$ and $b_\mathrm{HI,c2} = 39_{-5}^{+6} \ \mathrm{km/s}$. The \ion{H}{I} component immediately redward of the central component is constrained to be at $\varv_\mathrm{c,c3} = 51_{-10}^{+6} \ \mathrm{km/s}$, with $\log(N_\mathrm{HI,c3}/\mathrm{cm}^{-2}) = 13.4 \pm 0.1$ and $b_\mathrm{HI,c3} = 30_{-4}^{+5} \ \mathrm{km/s}$. Finally, the reddest \ion{H}{I} component is derived at $\varv_\mathrm{c,c3} = 66_{-9}^{+16} \ \mathrm{km/s}$, with $\log(N_\mathrm{HI,c4}/\mathrm{cm}^{-2}) = 12.8_{-0.6}^{+0.3}$ and $b_\mathrm{HI,c4} = 34_{-5}^{+6} \ \mathrm{km/s}$.

For the PIE modeling of this absorber, given the limited number of available ions, we use a single gas phase with density $\log(n_\mathrm{H}/\mathrm{cm}^{-3}) \approx -4.2$ constrained using \ion{O}{IV} and \ion{O}{V}. Despite not using \ion{O}{VI} in the fitting, the best-fit PIE model only slightly underproduces the measured column density, so it may be produced in the same gas phase as lower ions given consistent linewidths. The inferred metallicity is $[\mathrm{\alpha/H}] \approx -0.8$ while the relative carbon abundance is $[\mathrm{C/\alpha}] \approx -0.5$. The inferred cloud size is quite large, at $\log(l/\mathrm{kpc}) \approx 1.5$. The measured linewidths suggest that this cloud is dominated by non-thermal broadening, imposing an upper limit on gas temperature that is consistent with the PIE inferred temperature. This absorber also does not pose a problem of rapid cooling, so we do not perform TDP modeling for this system.

# z=1.26

For this absorber, we find significant detections of \ion{H}{I}, \ion{C}{III}, \ion{C}{IV}, and \ion{O}{IV}. The HIRES \ion{C}{IV} has a single narrow component with $b_\mathrm{CIV} \approx 7 \ \mathrm{km/s}$ at $\varv_\mathrm{c} \approx 0 \ \mathrm{km/s}$. While \ion{O}{IV} is also described by a single component, given the narrow absorption features, we are unable to constrain the linewidth independently, so we tie the value to $b_\mathrm{CIV}$ during fitting. For \ion{C}{III}, we find it necessary to apply an offset of $\delta \varv = 5 \ \mathrm{km/s}$ while also tying the linewidth to $b_\mathrm{CIV}$ during fitting.

To describe the absorption profiles for \ion{H}{I} 1215/1025/972, we find it necessary to invoke four components. c1 is coincident with the metal absorption, while c2, c3, and c4 (going redward) are metal-free. Given the limited number of \ion{H}{I} transitions, we need to impose several restrictions to derive significant constraints on each \ion{H}{I} component. First, we require $-80 \ \mathrm{km/s} \! \le \! \varv_\mathrm{c,c2} \!\le\! -50 \ \mathrm{km/s}$, $-50 \ \mathrm{km/s} \!\le\! \varv_\mathrm{c,c3} \!\le\! 15 \ \mathrm{km/s}$, and $15 \ \mathrm{km/s} \!\le\! \varv_\mathrm{c,c4} \!\le\! 60 \ \mathrm{km/s}$. Additionally, we also require $b_\mathrm{HI,c1} \!>\! b_\mathrm{HI,c2} \!>\! b_\mathrm{HI,c3} \!>\! b_\mathrm{HI,c4}$. We also require $b_\mathrm{HI,c4} \!>\! 6 \ \mathrm{km/s}$ to avoid c4 from becoming unreasonably narrow. During the simultaneous fitting, we further impose $b_\mathrm{HI,c1} \!<\! \sqrt{m_\mathrm{C}/m_\mathrm{H}} b_\mathrm{CIV,c1}$, which imposes an upper limit on the \ion{H}{I} linewidth, assuming that the entire broadening is thermal. Lastly, we require $N_\mathrm{HI,c1} \!>\! N_\mathrm{HI,c4}$.

We present the simultaneous Voigt profile-fitting results in the top panel of Figure \ref{fig:z_126} and Table \ref{fig:z_126}. The bluest metal-free \ion{H}{I} component is constrained to be at $\varv_\mathrm{c,c2} = -65_{-7}^{+5} \ \mathrm{km/s}$, with $\log(N_\mathrm{HI,c2}/\mathrm{cm}^{-2}) = 13.8 \pm 0.1$ and $b_\mathrm{HI,c2} = 20 \pm 2 \ \mathrm{km/s}$. The \ion{H}{I} component immediately blueward of the central component is constrained to be at $\varv_\mathrm{c,c3} = -38 \pm 5 \ \mathrm{km/s}$, with $\log(N_\mathrm{HI,c3}/\mathrm{cm}^{-2}) = 13.6 \pm 0.2$ and $b_\mathrm{HI,c3} = 17 \pm 3 \ \mathrm{km/s}$. Finally, the reddest \ion{H}{I} component is derived at $\varv_\mathrm{c,c3} = 27_{-6}^{+8} \ \mathrm{km/s}$, with $\log(N_\mathrm{HI,c4}/\mathrm{cm}^{-2}) = 12.5_{-0.3}^{+0.2}$ and $b_\mathrm{HI,c4} = 12 \pm 4 \ \mathrm{km/s}$.

For the PIE modeling, we constrain a single gas phase of density $\log(n_\mathrm{H}) \approx -3.3$ using \ion{C}{III} and \ion{C}{IV}. The metallicity estimate using \ion{H}{I} and \ion{O}{IV} yields $[\mathrm{\alpha/H}] \approx 0.6$; the high metallicity is a consequence of a low $N_\mathrm{HI,c1}$. The carbon elemental abundance is $[\mathrm{C/O}] \approx 0$ while the cloud size is small at $\log(l/\mathrm{kpc}) \approx -1.9$. However, the best-fit PIE model for this absorber has two problems. First, the gas temperature of $\log(T/\mathrm{K}) \approx 4.4$ is inconsistent with the upper limit on PIE temperature of 4.1 dex. Furthermore, the high metallicity of this system makes the cooling timescale shorter than the recombination timescale. This motivates us to explore TDP modeling for this system.

For this absorber, both inferred density and metallicity under TDP are consistent with PIE, while gas temperatures are allowed to be warmer, consistent with the linewidth-based inference. The inferred carbon abundance $[\mathrm{C/\alpha}] \approx 0.3$ is however larger than the PIE estimate, and this may be understood in terms of the underproduction of \ion{C}{IV} in TDP at this density range, as depicted in Figure \ref{fig:TDP_frac}. The inferred cloud size also remains consistent with PIE.


# z=1.27

For this absorber, we identify significant detection for \ion{H}{I}, \ion{C}{III}, \ion{C}{IV}, \ion{O}{III}, \ion{O}{IV}, \ion{O}{V}, \ion{O}{VI}, \ion{Ne}{V}, and \ion{Ne}{VI}. Both \ion{C}{IV} and \ion{O}{IV} reveal a two-component structure, with c1 at $\varv_c \approx 0 \ \mathrm{km/s}$ and c2 at $\varv_c \approx 40 \ \mathrm{km/s}$. By applying velocity offsets of $\delta \varv_c = 3.2, 2.5, 5.2 \ \mathrm{km/s}$, \ion{C}{III} 977, \ion{N}{IV} 765, and \ion{O}{III} 702 can be aligned with c1. During the fitting for \ion{C}{III}, we tie its linewidth to $b_\mathrm{CIV,c1}$. \ion{O}{V} also exhibits a two-component structure but requires a velocity offset of $\delta \varv = 15 \ \mathrm{km/s}$ to align it with \ion{C}{IV} and \ion{O}{IV}. Given the saturation of \ion{O}{V}, during fitting, we tie its linewidths to \ion{O}{IV} for each component and also restrict $N_\mathrm{OV,c1}\!>\!N_\mathrm{OV,c2}$. After applying an offset of $\delta \varv = 15 \ \mathrm{km/s}$, \ion{O}{VI} 1031 exhibits a two-component structure coinciding with c1 and c2. \ion{O}{VI} 1037 was contaminated with Ly$\alpha$ at $z=0.94481$. Given the saturation of the contaminating feature, while absorption for \ion{O}{VI} 1037 consistent with \ion{O}{VI} 1031 was recovered, absorption at c2 for \ion{O}{VI} 1037 could not be recovered. \ion{Ne}{V} 568 also reveals a two-component structure, although a velocity offset of $\delta \varv = -4.6 \ \mathrm{km/s}$ and a restriction of $N_\mathrm{NeV, c1}\!>\!N_\mathrm{NeV, c2}$ is necessary. Finally, \ion{Ne}{VI} 558 exhibits a two-component structure consistent with \ion{C}{IV} and \ion{O}{IV} without applying any velocity offsets or restrictions, and since it has higher ionization potential than \ion{O}{V}, \ion{O}{VI}, and \ion{Ne}{V}, this justifies trying to align the two-component structure in these ions with \ion{C}{IV} and \ion{O}{IV}. To explain the \ion{H}{I} 1215/1025 absorption, we require three components. Two components coincide with c1 and c2 while the third, c3, is metal-free. During fitting, we restrict $N_\mathrm{HI,c1}\!>N_\mathrm{HI,c2}>\!N_\mathrm{HI,c3}$ to derive constraints on all \ion{H}{I} components. The joint-Voigt profile fitting results are shown in the top panel of Figure \ref{fig:z_127} and Table \ref{tab:z_127}. The metal-free \ion{H}{I} component is constrained to be at $\varv_\mathrm{c,c3} = 71_{-10}^{+7} \ \mathrm{km/s}$ with $\log(N_\mathrm{HI}/\mathrm{cm}^{-2}) = 13.8 \pm 0.1$ and $b_\mathrm{HI,c3} = 29_{-5}^{+6} \ \mathrm{km/s}$.

The PIE modeling of c1 and c2 reveals distinct ionizing conditions. For c1, we treat \ion{O}{V} as a lower limit, given its saturation. Given consistent linewidths for \ion{O}{IV}, \ion{O}{V}, and \ion{O}{VI}, we use them to constrain a low density phase of $\log(n_\mathrm{HI,c1H}/\mathrm{cm}^{-3}) \approx -4.3$. Note that given the broader linewidths for \ion{Ne}{V} and \ion{Ne}{VI}, we do not include them in the fitting, however, the best-fit PIE model predicts consistent column densities with measured values. The high-ionization phase underproduces \ion{O}{III}, necessitating a more dense, low-ionization phase of density $\log(n_\mathrm{H}/\mathrm{cm}^{-3}) \approx -2.9$, which is largely constrained by sensitive \ion{He}{I}, \ion{Si}{III}, and \ion{Si}{IV} upper limits. During the simultaneous two-phase PIE fit, we attribute the \ion{C}{IV} and \ion{N}{IV} absorption to the low-density phase, given their consistent linewidths with \ion{O}{IV}; note that matching the \ion{C}{IV} absorption with the high-ionization phase ends up producing all of \ion{C}{III} so that the relative carbon abundance for the low-ionization phase is only an upper limit. Through the two-phase fit, we derive consistent metallicities of $[\mathrm{\alpha/H}] \approx 0$ for both phases. The constrained elemental abundances for the high-ionization phase are $\mathrm{[C/\alpha]} \approx -0.2$ and $\mathrm{[N/\alpha]} \approx -0.6$. The low-ionization phase has a cloud size smaller than $\log(l_\mathrm{c1L}/\mathrm{kpc}) < -0.1$, while the high-ionization phase is more extended, with $\log(l_\mathrm{c1H}/\mathrm{kpc}) \approx 1.4$. 

There are a few issues with the two-phase PIE decomposition. First, given consistent linewidths for \ion{O}{III}, \ion{O}{IV}, and \ion{O}{V}, they should arise in the same phase, but PIE cannot accommodate \ion{O}{III} in the high-ionization phase. Furthermore, the linewidth-based temperature for the low-ionization phase, combining \ion{H}{I} and \ion{O}{III}, is higher than the PIE-inferred temperature at about a $2 \sigma$ level. Finally, \ion{C}{III}, despite having a low-ionization potential, is entirely produced by the high-ionization phase. These considerations motivate using TDP to model this absorption component.

Using a diagnostic plot like that shown in Figure \ref{fig:TDP_diagnostic}, we find that even in TDP, \ion{O}{III}, \ion{O}{IV}, and \ion{O}{VI} cannot be accommodated in the same phase, although TDP can produce more \ion{O}{III} in the high-ionization phase relative to PIE. We therefore perform a two-phase TDP fit, similar to PIE. The inferred gas densities for both phases are consistent with PIE. While the metallicity for the high-ionization phase is consistent with PIE, the low-ionization phase has a lower metallicity of $[\mathrm{\alpha/H}] \approx -0.6$ relative to PIE. This can be understood in terms of the warmer gas temperatures permitted under TDP relative to PIE. The carbon relative abundance of the high-ionization phase $[\mathrm{C/\alpha}] \approx 0.4$ is enhanced relative to PIE, which can be attributed to the underproduction of \ion{C}{IV} in TDP in the density regime of $n_\mathrm{H} \approx 10^{-4} \ \mathrm{cm}^{-3}$ as demonstrated in Figure \ref{fig:TDP_frac}. The high-ionization phase still produces all of the measured \ion{C}{III} column density, resulting in upper limits for the relative carbon abundance of the low-ionization phase, $\mathrm{[C/\alpha]}<0$. The inferred cloud sizes under TDP are larger than PIE, with $\log(l_\mathrm{c1L}/\mathrm{kpc}) \approx -0.3$ and $\log(l_\mathrm{c1H}/\mathrm{kpc}) < 2.0$. Note that we do not incorporate \ion{Ne}{V} or \ion{Ne}{VI} in the fitting given their broader linewidths, but the high-ionization phase produces \ion{Ne}{V} consistent with the measurement while overproducing \ion{Ne}{VI}; the latter may be a consequence of overionization in TDP because of a recombination lag.

For the PIE modeling of c2, a single gas phase of density $\log(n_\mathrm{H,c2}/\mathrm{cm}^{-3}) \approx -4.1$ driven by \ion{O}{IV}, \ion{O}{V} (lower limit), and \ion{O}{VI} is sufficient. Note that given a single available transition for \ion{Ne}{V} and \ion{Ne}{VI}, we do not include them in fitting, however, the best-fit PIE model produces \ion{Ne}{V} consistent with the measured value while underproducing \ion{Ne}{VI}. The metallicity of this phase is $[\mathrm{\alpha/H}]_\mathrm{c2} \approx -0.4$ while the relative carbon abundance, driven by \ion{C}{IV}, is $[\mathrm{C/\alpha}]_\mathrm{c2} \approx -0.2$. The inferred cloud size, $\log(l_\mathrm{c2}/\mathrm{kpc}) \approx 1.0$, is large. The PIE-inferred temperature is smaller than the linewidth-based temperature (obtained using \ion{H}{I}, \ion{C}{IV}, \ion{O}{IV}, and \ion{O}{VI} linewidths), but only at a $1.5 \sigma$ level. Furthermore, the gas cools on a timescale larger than ionic recombination. Therefore, we do not pursue TDP modeling for this absorption component.
