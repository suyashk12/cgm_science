```python
import numpy as np
import pandas as pd
import tabula as tb
from tabulate import tabulate
```

```python
def df_concat_replace(df_left, df_right):
    
    df = pd.concat([df_left, df_right], ignore_index=True)
    df['Line'] = df['Line'].str.replace(' ','')
    df['W_r'] = df['W_r'].str.replace(' ','')
    df['logN'] = df['logN'].str.replace(' ','')
    df['b'] = df['b'].str.replace(' ','')
    df['v'] = df['v'].str.replace(' ','')
    df['v'] = df['v'].str.replace('−','-')
    df = df.reindex()
    
    return df
```

```python
def df_process(df):
    
    W_r_lo = []
    W_r_mid = []
    W_r_hi = []

    logN_lo = []
    logN_mid = []
    logN_hi = []

    b_lo = []
    b_mid = []
    b_hi = []

    v_lo = []
    v_mid = []
    v_hi = []

    for i in range(len(df)):
                
        line = df.loc[i,'Line']
        W_r = df.loc[i,'W_r']
        logN = df.loc[i,'logN']
        b = df.loc[i,'b']
        v = df.loc[i,'v']

        if type(line) == float:
            df.loc[i,'Line'] = df.loc[i-1,'Line']

        if type(W_r) == float:
            W_r_lo.append(np.nan)
            W_r_mid.append(np.nan)
            W_r_hi.append(np.nan)

        elif type(W_r) == str:
            if W_r[0] == '>':
                W_r_lo.append(float(W_r[1:]))
                W_r_mid.append(np.nan)
                W_r_hi.append(np.nan)

            elif W_r[0] == '<':
                W_r_lo.append(np.nan)
                W_r_mid.append(np.nan)
                W_r_hi.append(float(W_r[1:]))

            else:
                W_r_lo.append(float(W_r.split('±')[0])-float(W_r.split('±')[1]))
                W_r_mid.append(float(W_r.split('±')[0]))
                W_r_hi.append(float(W_r.split('±')[0])+float(W_r.split('±')[1]))

        if type(logN) == float:
            logN_lo.append(np.nan)
            logN_mid.append(np.nan)
            logN_hi.append(np.nan)

        elif type(logN) == str:
            if logN[0] == '>':
                logN_lo.append(float(logN[1:]))
                logN_mid.append(np.nan)
                logN_hi.append(np.nan)

            elif logN[0] == '<':
                logN_lo.append(np.nan)
                logN_mid.append(np.nan)
                logN_hi.append(float(logN[1:]))

            else:
                logN_lo.append(float(logN.split('±')[0])-float(logN.split('±')[1]))
                logN_mid.append(float(logN.split('±')[0]))
                logN_hi.append(float(logN.split('±')[0])+float(logN.split('±')[1]))

        if type(b) == float:
            b_lo.append(np.nan)
            b_mid.append(np.nan)
            b_hi.append(np.nan)

        elif type(b) == str:
            if b[0] == '>':
                b_lo.append(float(b[1:]))
                b_mid.append(np.nan)
                b_hi.append(np.nan)

            elif b[0] == '<':
                b_lo.append(np.nan)
                b_mid.append(np.nan)
                b_hi.append(float(b[1:]))
                
            elif '±' not in b:
                b_lo.append(np.nan)
                b_mid.append(float(b.split('±')[0]))
                b_hi.append(np.nan)

            else:
                b_lo.append(float(b.split('±')[0])-float(b.split('±')[1]))
                b_mid.append(float(b.split('±')[0]))
                b_hi.append(float(b.split('±')[0])+float(b.split('±')[1]))

        if v[0] == '[':
            v_lo.append(float(v[1:-1].split(',')[0]))
            v_mid.append(.5*(float(v[1:-1].split(',')[0])+float(v[1:-1].split(',')[1])))
            v_hi.append(float(v[1:-1].split(',')[1]))

        else:
            v_lo.append(float(v.split('±')[0])-float(v.split('±')[1]))
            v_mid.append(float(v.split('±')[0]))
            v_hi.append(float(v.split('±')[0])+float(v.split('±')[1]))        


    df['W_r_lo_ss'] = W_r_lo
    df['W_r_mid_ss'] = W_r_mid
    df['W_r_hi_ss'] = W_r_hi

    df['logN_lo_ss'] = logN_lo
    df['logN_mid_ss'] = logN_mid
    df['logN_hi_ss'] = logN_hi

    df['b_lo_ss'] = b_lo
    df['b_mid_ss'] = b_mid
    df['b_hi_ss'] = b_hi

    df['v_lo_ss'] = v_lo
    df['v_mid_ss'] = v_mid
    df['v_hi_ss'] = v_hi

    df = df.drop(columns=['W_r','logN','b','v'])    
    
    return df
```
