Define function for constructing a Voigt profile

```python
def vp(v, logN, lsf_convolve=True):
    
    model = comp_model_spec_gen(v, np.array([[logN, b, 0]]), 
                                cii_1036_fuv.wav0_rest, cii_1036_fuv.f, cii_1036_fuv.gamma, cii_1036_fuv.A,
                                lsf_convolve,
                                cii_1036_fuv.lsf, cii_1036_fuv.v_lsf)[1]
    
    return model
```

Construct a mock profile to test convolution and measure the new FWHM

```python
# Unconvolved and convolved profiles
mock_unconv = vp(v, 14.5, False)
mock_conv = vp(v, 14.5, True)

# Flux levels for FWHM in each profile
flux_fwhm_unconv = 1-0.5*(1-np.min(mock_unconv))
flux_fwhm_conv = 1-0.5*(1-np.min(mock_conv))

# Corresponding FWHM calculation
fwhm_unconv = 2*v[v>0][np.argmin(np.abs(mock_unconv[v>0]-flux_fwhm_unconv))]
fwhm_conv = 2*v[v>0][np.argmin(np.abs(mock_conv[v>0]-flux_fwhm_conv))]
```

```python
idx = (v>-fwhm_conv)&(v<fwhm_conv)
err_std = cii_1036_fuv.err_norm
```

Run the Monte Carlo experiment

```python
N_trials = 10000
logN_grid = np.linspace(12.5, 14.0, 50)
```

```python
# For each logN for each trial, get Gaussian noise
noise_grid = np.zeros((len(logN_grid), N_trials, len(v)))
```

```python
# For each spectral pixel
for k in range(len(v)):
    # Generate noise for all logN, for each trial
    noise_samples = np.random.normal(loc=0, scale=err_std[k], size=(len(logN_grid), N_trials))
    # Save the reshaped noise in the larger grid
    noise_grid[:,:,k] = noise_samples
```

```python
# For each logN, generate a model
model_grid = np.zeros((len(logN_grid), len(v)))
# And for each trial, add noise to the model
flux_grid = np.zeros((len(logN_grid), N_trials, len(v)))

for i in range(len(logN_grid)):
    model_grid[i,:] = vp(v, logN_grid[i])
    
    for j in range(N_trials):
        flux_grid[i,j,:] = model_grid[i,:]+noise_grid[i,j,:]
```

Compare EW and $\chi^2$ for the models

```python
from scipy.stats import ks
```

```python
chi_sq_model_grid = np.zeros((len(logN_grid)))
EW_model_grid = np.zeros(len(logN_grid))
```

```python
def logN_2_EW(logN):
    
    return (cii_1036_fuv.wav0_rest*1e+3)*((cii_1036_fuv.wav0_rest*1e-10)/3e+8)*(2.654e-2*cii_1036_fuv.f)*10**logN
```

```python
delta_lambda = np.mean(((v[1:]-v[:-1])*cii_1036_fuv.wav0_rest/3e+5)*1e+3) # in mA
```

```python
for i in range(len(logN_grid)):
    
    model = model_grid[i]
    
    chi_sq_model_grid[i] = np.sum((model[idx]-1)**2/err_std[idx]**2)
    EW_model_grid[i] = np.sum((1-model[:-1]))
```

```python
x = np.linspace(0,1,100)
y=np.exp(-x)
```

```python
fig, ax = plt.subplots(1, figsize=(5,4))

ax.plot(logN_grid, np.log10(np.sqrt(chi_sq_model_grid)), label=r'$\chi^2$')
#ax.plot(logN_grid, logN_2_EW(logN_grid), label=r'$W$ (m, curve of growth)')
ax.plot(logN_grid, np.log10(EW_model_grid), label=r'$W$ (direct)')

#ax.set_xlabel(r'$N$')
ax.set_xlabel(r'$\log N$')
ax.set_ylabel('Statistic')
ax.legend()
```

The departure between direct measurement and curve of growth inference of $W$ is because of the breakdown of the approximation -

$$\exp(-x) \approx 1-x$$

```python
tau_nu_grid = np.zeros(len(logN_grid))
for i in range(len(logN_grid)):
    tau_nu_grid[i] = np.max(-np.log(model_grid[i,:]))
```

```python
fig, ax = plt.subplots(1, figsize=(5,4))
ax.plot(logN_grid, np.exp(-tau_nu_grid), label=r'$\exp(-\tau_\nu)$')
ax.plot(logN_grid, 1-tau_nu_grid, label=r'$1-\tau_\nu$')
ax.set_xlabel(r'$\log N$')
ax.set_ylabel(r'$y$')
ax.legend()
```

Calculate the chi-squared value for all trials -

$$\chi_0^2 = \sum_{i=1}^k \frac{(f_i-1)^2}{\sigma_f^2}$$

```python
chi_sq_grid = np.zeros((len(logN_grid), N_trials))
chi_sq_grid_null = np.zeros((len(logN_grid), N_trials))
```

```python
for i in range(len(logN_grid)):
    for j in range(N_trials):
        chi_sq_grid[i,j] = np.sum((flux_grid[i,j,:][idx]-1)**2/err_std[idx]**2)
        chi_sq_grid_null[i,j] = np.sum((noise_grid[i,j,:][idx])**2/err_std[idx]**2)
```

Calculate the equivalent widths for all trials -

```python
EW_grid = np.zeros((len(logN_grid), N_trials))
EW_grid_null = np.zeros((len(logN_grid), N_trials))
```

```python
delta_lambda = np.mean(((v[1:]-v[:-1])*cii_1036_fuv.wav0_rest/3e+5)*1e+3) # in mA
```

```python
for i in range(len(logN_grid)):
    for j in range(N_trials):
        EW_grid[i,j] = np.sum((1-flux_grid[i,j,:][idx][:-1])*delta_lambda)
        EW_grid_null[i,j] = np.sum((noise_grid[i,j,:][idx][:-1])*delta_lambda)
```

Make some plots to check

Compute the empirical CDF for $\chi^2$

```python
N_bins_chi_sq = 100
CDF_chi_sq_interp = []
```

```python
for i in range(len(logN_grid)):
    
    chi_sq_samples_null = chi_sq_grid_null[i,:]
    hist, bins = np.histogram(chi_sq_samples_null, bins=N_bins_chi_sq, density=True)
    cdf = np.cumsum(hist)/np.sum(hist)
    
    CDF_chi_sq_interp.append(interpolate.interp1d(x=.5*(bins[1:]+bins[:-1]),
                            y=cdf, fill_value='extrapolate'))
```

```python
p_values_chi_sq_grid = 1-CDF_chi_sq_interp[i](chi_sq_grid)
```

```python
p_values_chi_sq_median = np.median(p_values_chi_sq_grid, axis=1)
p_values_chi_sq_inv_interp = interpolate.interp1d(x=p_values_chi_sq_median, y=logN_grid, fill_value='extrapolate')
```

```python
plt.plot(logN_grid, p_values_chi_sq_median)
```

```python
# Sigma values taken from Gehrels 1986

cdf_1sig = 0.8413
cdf_2sig = 0.9772
cdf_3sig = 0.9987
```

```python
logN_chi_sq_1sig = p_values_chi_sq_inv_interp([1-cdf_1sig])[0]
logN_chi_sq_2sig = p_values_chi_sq_inv_interp([1-cdf_2sig])[0]
logN_chi_sq_3sig = p_values_chi_sq_inv_interp([1-cdf_3sig])[0]
```
