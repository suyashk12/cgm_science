Understanding a voigt profile as the convolution of a Gaussian and a Lorentzian

```python
def gauss(x, mu, sigma):
    
    return (1/np.sqrt(2*np.pi*sigma**2))*np.exp(-(x-mu)**2/(2*sigma**2))
```

```python
def lorentz(x, mu, gamma):
    
    return (1/(np.pi*gamma))*(gamma**2/((x-mu)**2 + gamma**2))
```

We perform a discrete convolution of a Gaussian and Lorentzian below. We use the mode "same" from <code>numpy.convolve()</code> to return a convolved profile under the "same" mode (which performs a "full" convolution and only returns the central results)

```python
def voigt(x, mu, sigma, gamma):
    
    G = gauss(x, mu, np.float_(sigma))
    L = lorentz(x, mu, gamma)
        
    # Multiply dx to convert discrete convolution to continuous
    return np.convolve(G, L, mode='same')*np.abs(np.mean(x[1:]-x[:-1]))
```

```python
fig, ax = plt.subplots(1, figsize=(15,7))

ax.scatter(x_mod, voigt_profile(x_mod, 1e+11, 1e+8), label='High sampling rate')
ax.scatter(x, voigt_profile(x, 1e+11, 1e+8), label='Regular sampling rate')

ax.set_xlabel('Relative frequency (Hz)')
ax.set_ylabel('Line profile')

ax.legend()
```

So, not only is the scipy solution faster, it also removes the necessity of sampling too frequently! Let's just use that to generate our optical depth.

```python
def tau_nu(x, N, sigma):
    
    # Oscillator strength and intrinsic damping taken from plotabs
    
    f = 0.079120  
    gamma = 1.897E+8
        
    return N*2.654E-2*f*voigt_profile(x, sigma, gamma/(4*np.pi))
```

```python
y_mod = np.exp(-tau_nu(x_mod, 10**14.4, 10**11.3))

# This model is just for testing if LSF convolution is actually working
#y_mod = np.exp(-tau_nu(x_mod, 10**10.5, 10**8))
```

```python
plot_idx_mod_new = (v_mod>-205) & (v_mod<205)

x_mod_new = del_nu_mod[plot_idx_mod_new]
```

```python
y_mod_new = np.exp(-tau_nu(x_mod_new, 10**14.4, 10**11.3))
```

```python
y_mod_conv_new = np.convolve(lsf_arr, y_mod_new, mode='same')
```

```python
n_trim = 50
```

```python
wav_mod_trim = wav_mod[plot_idx_mod_new][n_trim:-n_trim]
y_mod_trim = y_mod_conv_new[n_trim:-n_trim]
```

```python
fig, ax = plt.subplots(1, figsize=(15,7))

ax.scatter(wav[plot_idx], y, color='black', label='Data')
ax.errorbar(wav[plot_idx], y, yerr = dy, color='black', linestyle='', lw=1)
ax.scatter(wav[plot_idx], y_mod_interp, label='Model')

ax.set_xlabel('Wavelength ()')
ax.set_ylabel('Continuum normalized flux')

#ax.set_xlim(1353.47, 1353.472)
ax.legend()
```

We are ready to perform our optimization now to find the Gaussian width and the column density, but first, let us write a convenience function that only takes as its input the list of relevant wavelengths along with a column density and a Gaussian width, spitting out the model spectrum sampled at exactly the input wavelengths

```python
# Could do better in terms of coding etiquette to not use things defined outside the function

def model(wavs_line, N, sigma):
        
    y_mod_new = np.exp(-tau_nu(x_mod_new, N, sigma))
        
    y_mod_conv_new = np.convolve(lsf_arr, y_mod_new, mode='same')
    
    y_mod_trim = y_mod_conv_new[n_trim:-n_trim]
    
    y_mod_interp = np.interp(wavs_line, wav_mod_trim, y_mod_trim)
    
    return y_mod_interp
```

Just testing the function out here

```python
fig, ax = plt.subplots(1, figsize=(15,7))

ax.scatter(wav[plot_idx], y, color='black', label='Data')
ax.errorbar(wav[plot_idx], y, yerr = dy, color='black', linestyle='', lw=1)
ax.scatter(wav[plot_idx], model(wav[plot_idx], 10**14.4, 10**11.3), label='Model 1')
ax.scatter(wav[plot_idx], model(wav[plot_idx], 10**14.25, 10**11.7), label='Model 2')

ax.set_xlabel('Wavelength ()')
ax.set_ylabel('Continuum normalized flux')

#ax.set_xlim(1353.47, 1353.472)
ax.legend()
```

Now onto LMFIT, let us get the column density and Gaussian width alongside their uncertainties

```python
vmod = Model(model)
result = vmod.fit(y, wavs_line=wav[plot_idx], N=10**14.4, sigma=10**11.3, weights=1/dy)
```
