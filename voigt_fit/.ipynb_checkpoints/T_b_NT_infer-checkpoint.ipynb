```python
# Significance levels
# Sigma values taken from Gehrels 1986
cdf_1sig = 0.8413
cdf_2sig = 0.9772
cdf_3sig = 0.9987
```

```python
# Boltzmann constant, adjusted to use km instead of m
k_B = 1.380649e-29
# Mass of hydrogen atom
amu = 1.66054e-27
```

# Functions

```python
def gen_b(logT, b_NT, A):
    
    return np.sqrt((2*k_B*10**logT/(A*amu)) + b_NT**2)
```

```python
def log_prior(params):

    # Grid parameters being varied
    logT, b_NT = params
    
    # Avoid edges?
    if 4<logT<6 and 0<b_NT<50:
        return 0.0
    return -np.inf
```

```python
def log_likelihood(params, b_list):
    
    logT, b_NT = params

    ll = 0
    
    for i in range(len(b_list)):
        
        ybar = gen_b(logT, b_NT, A_dict[b_list[i][1]])
        y = b_list[i][2]
        dy = b_list[i][3]
        
        ll += (y-ybar)**2/dy**2
        
    return ll
```

```python
def log_probability(params, b_list):
        
    lp = log_prior(params)
    
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(params, b_list)
```

```python
b_list = [[['HI', 'H', 34.5, 1.7],
          ['CIII', 'C', 32.3, 2.0],
          ['CIV', 'C', 36.7, 6.1],
          ['OIII', 'O', 34.3, 1.9],
          ['OIV', 'O', 30.2, 1.8]]]
```

```python
pos = np.array([4.5, 32]) + 1e-4 * np.random.randn(500, 2)
nwalkers, ndim = pos.shape
```

```python
sampler = emcee.EnsembleSampler(
    nwalkers, ndim, log_probability, args=(b_list)
)
```

```python
sampler.run_mcmc(pos, 500, progress=True)
```
