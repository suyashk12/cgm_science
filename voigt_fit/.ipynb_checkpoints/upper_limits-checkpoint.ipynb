```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import voigt_profile
from scipy import integrate, interpolate
from lmfit import minimize, Parameters
import emcee
import corner
```

Specify atomic parameters for LyA

```python
wav0_rest = 1215.6701
f = 0.416400
gamma_nu = 6.265E8
```

Convert $\Gamma_\nu$ to $\Gamma_v$ (must be done in the rest frame)

```python
# First convert wavelength to meters, then velocity to km/s
gamma_v = (gamma_nu*(wav0_rest*1e-10))*1e-3
```

Thomson scattering cross-section

```python
sigma_0 = 2.654E-2 # In cm^-2
```

Specify the redshift, wavelength range of observation, and velocity pixel size

```python
z_test = 0.1
wav0_obs_min = 1250
wav0_obs_max = 1450
delta_v_pix = 2.3
```

Use this information to generate the velocity array

```python
wav0_obs = wav0_rest*(1+z_test) # Observed wavelength of transition
v_min = 3e+5*(wav0_obs_min-wav0_obs)/wav0_obs
v_max = 3e+5*(wav0_obs_max-wav0_obs)/wav0_obs
v = np.arange(v_min, v_max+delta_v_pix, delta_v_pix)
```

Generate the Ly$\alpha$ lines

```python
N = 1e+12 # cm^-2
b = 12.9
```

```python
def vp(v, N):
    
    sigma_v = b/np.sqrt(2)
    tau = N*sigma_0*f*voigt_profile(v, sigma_v, gamma_v/(4*np.pi))*wav0_rest*1e-13
    flux = np.exp(-tau)
    
    return flux
```

```python
flux = vp(v, N)
```

```python
fig, ax = plt.subplots(1, figsize=(7,4))
ax.step(v, flux, where='mid')

ax.axhline(0, linestyle=':', color='red')
ax.axhline(1, linestyle=':', color='green')
ax.axvline(0, linestyle=':', color='brown')


ax.set_xlim(-100,100)
ax.set_ylim(-0.2,1.2)

ax.set_xlabel('Relative velocity (km/s)')
ax.set_ylabel('Continuum normalized flux')
```

Create the noisy data

```python
err_std = 0.15
# Draw samples from a Gaussian of mean zero and above std dev
np.random.seed(45)
pert = np.random.normal(loc=0, scale=err_std, size=len(v))
flux_pert = flux+pert
err_pert = np.ones(len(v))*err_std
```

```python
fig, ax = plt.subplots(1, figsize=(7,4))
ax.step(v, flux_pert, where='mid')
ax.step(v, err_pert, color='cyan', where='mid')

ax.axhline(0, linestyle=':', color='red')
ax.axhline(1, linestyle=':', color='green')
ax.axvline(0, linestyle=':', color='brown')


ax.set_xlim(-100,100)
ax.set_ylim(-0.05,1.5)

ax.set_xlabel('Relative velocity (km/s)')
ax.set_ylabel('Continuum normalized flux')
```

Calculate EW within a window equal to twice of the FWHM

```python
fwhm = 2*np.sqrt(2*np.log(2))*sigma_v
```

```python
idx = (v>-fwhm)&(v<fwhm)
```

```python
fig, ax = plt.subplots(1, figsize=(7,4))
ax.step(v, flux_pert, where='mid')

ax.step(v, err_std*np.ones(len(v)), color='cyan', where='mid')
ax.axvspan(xmin=-fwhm, xmax=fwhm, color='yellow', alpha=.1)

ax.axhline(0, linestyle=':', color='red')
ax.axhline(1, linestyle=':', color='green')
ax.axvline(0, linestyle=':', color='brown')


ax.set_xlim(-100,100)
ax.set_ylim(-0.05,1.4)

ax.set_xlabel('Relative velocity (km/s)')
ax.set_ylabel('Continuum normalized flux')
```

```python
# Calculate EW of the unperturbed and perturbed feature in mÅ

EW = np.sum(1-flux[idx][:-1])*(delta_v_pix*wav0_rest/3e+5)*1e+3
EW_pert = np.sum(1-flux_pert[idx][:-1])*(delta_v_pix*wav0_rest/3e+5)*1e+3
sig_EW_pert = np.sqrt(np.sum(err_pert[idx][:-1]**2))*(delta_v_pix*wav0_rest/3e+5)*1e+3
```

```python
EW, EW_pert, sig_EW_pert
```

Fit a Voigt profile using LMFIT

```python
def residual(param, v, flux, err):
    
    model = vp(v, 10**param['logN'].value)
    
    return (flux-model)/err
```

```python
params = Parameters()
#params.add('N', value=1e+12, min=0)
params.add('logN', value=12)
```

```python
result = minimize(residual, params, args=(v, flux_pert, err_pert))
```

Refine using MCMC

```python
def log_likelihood(logN, v, flux, err):
    
    model = vp(v, 10**logN)
    return -0.5*np.sum(((flux-model)/err)**2)
```

```python
def log_prior(logN):
    if logN>-10 and logN<18.5:
        return 0
    else:
        return -np.inf
```

```python
def log_probability(logN, v, flux, err):
    lp = log_prior(logN)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(logN, v, flux, err)
```

```python
pos = result.params['logN'].value + 1e-4 * np.random.randn(200, 1)
nwalkers, ndim = pos.shape
```

```python
sampler = emcee.EnsembleSampler(
    nwalkers, ndim, log_probability, args=(v, flux_pert, err_pert)
)
```

```python
sampler.run_mcmc(pos, 6000, progress=True)
```

```python
flat_samples = sampler.get_chain(discard=100, flat=True)
```

```python
pdf, bin_edges = np.histogram(flat_samples, bins=100, density=True)
X = 10**(.5*(bin_edges[1:]+bin_edges[:-1]))
pdf = pdf/integrate.trapz(x=X, y=pdf)
cdf = integrate.cumtrapz(x=X, y=pdf, initial=0)/integrate.trapz(x=X, y=pdf)
cdf_inv_interp = interpolate.interp1d(x=cdf, y=X)
```

```python
fig, ax = plt.subplots(1, figsize=(7,4))

ax.plot(X, pdf)
ax.axvline(cdf_inv_interp(0.5), linestyle=':')
ax.axvline(cdf_inv_interp(0.16), linestyle=':')

ax.set_xlabel(r'$N$')
ax.set_ylabel('PDF')
ax.set_xscale('log')
```
