Define function for constructing a Voigt profile

```python
def vp(v, logN, lsf_convolve=True):
    
    model = comp_model_spec_gen(v, np.array([[logN, b, 0]]), 
                                cii_1036_fuv.wav0_rest, cii_1036_fuv.f, cii_1036_fuv.gamma, cii_1036_fuv.A,
                                lsf_convolve,
                                cii_1036_fuv.lsf, cii_1036_fuv.v_lsf)[1]
    
    return model
```

Construct a mock profile to test convolution and measure the new FWHM

```python
# Unconvolved and convolved profiles
mock_unconv = vp(v, 12.5, False)
mock_conv = vp(v, 12.5, True)

# Flux levels for FWHM in each profile
flux_fwhm_unconv = 1-0.5*(1-np.min(mock_unconv))
flux_fwhm_conv = 1-0.5*(1-np.min(mock_conv))

# Corresponding FWHM calculation
fwhm_unconv = 2*v[v>0][np.argmin(np.abs(mock_unconv[v>0]-flux_fwhm_unconv))]
fwhm_conv = 2*v[v>0][np.argmin(np.abs(mock_conv[v>0]-flux_fwhm_conv))]
```

```python
fig, ax = plt.subplots(1, figsize=(7,4))
ax.plot(v, mock_unconv, label='Unconvolved', color='black')
ax.axhline(flux_fwhm_unconv, color='black', linestyle=':')
ax.plot(v, mock_conv, label='Convolved', color='purple')
ax.axhline(flux_fwhm_conv, color='purple', linestyle=':')
ax.legend()
```

```python
fig, ax = cii_1036_fuv.plot_ion_transition_spec()
ax.axvspan(xmin=-fwhm_conv, xmax=fwhm_conv, color='yellow', alpha=.1)
```

Create a grid of $logN$

```python
idx = (v>-fwhm_conv)&(v<fwhm_conv)
err_std = cii_1036_fuv.err_norm
```

```python
logN_grid = np.linspace(12.85, 14.3, 2000)
model_grid = np.array([vp(v, logN, lsf_convolve=True) for logN in logN_grid])
```

Run the Monte Carlo experiment

```python
N_trials = 2000
```

```python
noise_grid = np.zeros((N_trials, len(logN_grid), len(v)))
```

```python
for k in range(len(v)):
    noise_samples = np.random.normal(loc=0, scale=err_std[k], size=N_trials*len(logN_grid)).reshape((N_trials, len(logN_grid)))
    noise_grid[:,:,k] = noise_samples
```

```python
flux_grid = np.zeros((N_trials, len(logN_grid), len(v)))

for i in range(N_trials):
    for j in range(len(logN_grid)):
        flux_grid[i,j,:] = model_grid[j,:]+noise_grid[i,j,:]
```

Make some plots to check

```python
# Pick some logN to plot
logN_plot_grid = np.linspace(np.min(logN_grid), np.max(logN_grid), 4)
```

```python
# Pick a trial at random

fig, axes = plt.subplots(2, 2, figsize=(15,8), sharex=True, sharey=True)

for i in range(len(logN_plot_grid)):
    
    logN = logN_plot_grid[i]
    ax = axes[i//2, i%2]
    
    # Find the closest logN for which the trial was run
    sample_logN_idx = np.argmin(np.abs(logN_grid-logN))
    sample_trial_idx = np.random.randint(low=0, high=N_trials, size=1)[0]
    
    # Isolate the flux and model
    flux = flux_grid[sample_trial_idx, sample_logN_idx, :]
    model = vp(v, logN)
    
    # Calculate probabilities
    chi_sq_0 = np.sum((flux[idx]-1)**2/err_std[idx]**2)
    
    # Make the plot
    ax.step(v, flux, where='mid', lw=.5)
    
    ax.axhline(1, color='red')
    ax.axhline(0, color='black', linestyle=':')
    ax.axvline(0, color='black', linestyle=':')

    ax.plot(v, model, color='green')

    ax.step(v, err_std*np.ones(len(v)), where='mid', color='cyan')
    
    ax.axvspan(xmin=-fwhm_conv, xmax=fwhm_conv, color='yellow', alpha=.1)
    
    ax.text(x=-270, y=0.3, s=r'$\log N$ = '+'{:.2f}'.format(logN))
    ax.text(x=160, y=0.3, s=r'$\chi_0^2$ = '+'{:.2f}'.format(chi_sq_0))

    ax.set_xlim(-300,300)
    ax.set_ylim(-0.2, 1.5)
    
ax_label = fig.add_subplot(111, frameon=False)
plt.tick_params(labelcolor='none', top='off', bottom='off', left='off', right='off')

ax_label.set_xlabel('Velocity (km/s)')
ax_label.set_ylabel('Flux (continuum normalized)')
plt.subplots_adjust(wspace=0, hspace=0)
```

Calculate $\chi_0^2 = \sum_{i=1}^k \frac{(f_i-1)^2}{\sigma_f^2}$ for all trials

```python
chi_sq_grid = np.zeros((N_trials, len(logN_grid)))
```

```python
for i in range(N_trials):
    for j in range(len(logN_grid)):
        chi_sq_grid[i,j] = np.sum((flux_grid[i,j,:][idx]-1)**2/err_std[idx]**2)
```

```python
# Plot results from some Monte carlo runs

N_trials_plot = 2
trials_idx = np.random.randint(low=0, high=N_trials, size=N_trials_plot)

fig, ax = plt.subplots(1, figsize=(7,4))

for i in range(N_trials_plot):
    ax.plot(logN_grid, chi_sq_grid[trials_idx[i],:], lw=.2)

ax.set_xlabel(r'$\log N$')
ax.set_ylabel(r'$\chi_0^2$')
```

Create the $\chi^2$ probability distribution function and CDF

```python
def chi_sq_pdf(chi_sq, nu):
    return chi_sq**(0.5*nu-1)*np.exp(-0.5*chi_sq)
```

```python
x_arr = np.linspace(0, 110, 1000)
dof = len(v[idx])-1

pdf = np.array([chi_sq_pdf(x, dof) for x in x_arr])
pdf /= np.sum(pdf)
p_tail = 1-np.cumsum(pdf)/np.sum(pdf)

p_tail_interp = interpolate.interp1d(x=x_arr, y=p_tail, fill_value='extrapolate')
```

Try smoothing the median curve

```python
def smooth_curve(x, y, delta_x):
    
    x_min = x[0]
    # x where x is less than delta_x from the end
    x_max = x[x<(x[-1]-delta_x)][-1]
    
    # Number of evaluations
    N_windows = np.where(x==x_max)[0][0]
    
    # Smoothed points
    x_smooth = np.zeros(N_windows)
    y_smooth = np.zeros(N_windows)
    
    for i in range(N_windows):
        window_idx = (x>=x[i])&(x<x[i]+delta_x)
        x_window = x[window_idx]
        y_window = y[window_idx]
        
        x_smooth[i] = np.median(x_window)
        y_smooth[i] = np.median(y_window)
    
    return x_smooth, y_smooth
```

```python
logN_smooth, p_tail_smooth = smooth_curve(logN_grid, p_tail_medians, 0.02)
p_tail_logN_smooth_interp = interpolate.interp1d(x=p_tail_smooth, y=logN_smooth)

logN_1sig_smooth = p_tail_logN_smooth_interp(p_1sig)
logN_2sig_smooth = p_tail_logN_smooth_interp(p_2sig)
logN_3sig_smooth = p_tail_logN_smooth_interp(p_3sig)
```
