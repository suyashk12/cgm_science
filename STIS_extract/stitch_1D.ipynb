```python
from astropy.io import fits
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from stis_extract import *
```

# J011035

```python
for es in [3, 4, 5, 6, 7, 8]:
    
    # Obtain unstitched files for a given extraction box
    spec = spec_1d('J011035', es)
    spec.load_spec1d_unstitched()
    
    # Parse the unstitched files across all exposures
    spec.parse_spec1d_unstitched()
    
    # Stitch and save individual exposures
    for i in range(spec.n_exps):
        spec.stitch_spec1d([i], 
                      rootdir+'J011035/{}/'.format(spec.exp_names[i]), 
                      'J011035_{}_es={}'.format(spec.exp_names[i],es))
        
    # Combine all exposures too
    spec.stitch_spec1d(np.arange(spec.n_exps), rootdir+'J011035/', 'J011035_es={}'.format(es))
```
