# J011035

```python
spec = np.loadtxt(rootdir+'J011035/J011035_x1d_es=4.dat')
```

```python
wav = spec[:,0]
flux = spec[:,1]
err = spec[:,2]
```

```python
fig, ax = plt.subplots(1, figsize=(15,4))

ax.step(wav, flux, where='mid', color='black')
ax.step(wav, err, where='mid', color='cyan')

ax.set_xlabel('Wavelength (Å)')
ax.set_ylabel(r'Flux (erg/s/cm${}^2$/Å)')

ax.set_title('J011035 (extrsize=4)')
```

```python
def wav_2_z(w):
    return w/1548-1
```

```python
def plot_CIV_markers(ax, z):
    ax.axvline(1548*(1+z), linestyle='--', color='red')
    ax.axvline(1550*(1+z), linestyle='--', color='red')
```

```python
wav_2_z(2045)
```

```python
fig, ax = plt.subplots(1, figsize=(15,4))

ax.step(wav, flux, where='mid', color='black')
ax.step(wav, err, where='mid', color='cyan')

plot_CIV_markers(ax, 0.2957)
plot_CIV_markers(ax, 0.3193)

ax.set_xlim(1950,2000)

ax.set_xlabel('Wavelength (Å)')
ax.set_ylabel(r'Flux (erg/s/cm${}^2$/Å)')

ax.set_title('J011035 (extrsize=4)')
ax.set_ylim(-1e-14, 1.5e-14)
```
