```python
import stistools
import os
import glob
from pathlib import Path
```

```python
rootdir = '/Users/thepoetoftwilight/Documents/CUBS/Data/CONTACT/STIS/'
```

```python
def extract_1D(qso_name, extrsize=None):

    # Reference directory for finding calibration files
    os.environ['oref'] = rootdir+'{}/2D/'.format(qso_name)

    # Route to 2D files downloaded from MAST
    dir_list = glob.glob(rootdir+'{}/2D/*flt*'.format(qso_name))

    for f in dir_list:
        # Get exposure name
        exp_name = f.split('/')[-1].split('.')[0][:-4]

        # Create path
        if extrsize==None:
            es_str = ''
        else:
            es_str = '_es={}'.format(extrsize)

        # Make path available
        Path(rootdir+'{}/unstitched{}'.format(qso_name, es_str)).mkdir(parents=True, exist_ok=True)
        
        # Stitch and save 1D files
        stistools.x1d.x1d(f, 
                          output=rootdir+'{}/unstitched{}/{}_x1d'.format(qso_name, es_str, exp_name),
                          extrsize=extrsize) 
```
