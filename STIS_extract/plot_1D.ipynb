# J011035

```python
spec_def = np.loadtxt(rootdir+'J011035/J011035_x1d.dat')
spec_cust_1 = np.loadtxt(rootdir+'J011035/J011035_x1d_es=3.dat')
spec_cust_2 = np.loadtxt(rootdir+'J011035/J011035_x1d_es=4.dat')
spec_cust_3 = np.loadtxt(rootdir+'J011035/J011035_x1d_es=5.dat')
```

```python
wav_def, flux_def, err_def = spec_def[:,0], spec_def[:,1], spec_def[:,2]
wav_cust_1, flux_cust_1, err_cust_1 = spec_cust_1[:,0], spec_cust_1[:,1], spec_cust_1[:,2]
wav_cust_2, flux_cust_2, err_cust_2 = spec_cust_2[:,0], spec_cust_2[:,1], spec_cust_2[:,2]
wav_cust_3, flux_cust_3, err_cust_3 = spec_cust_3[:,0], spec_cust_3[:,1], spec_cust_3[:,2]
```

```python
fig, ax = plt.subplots(1, figsize=(15,4))
ax.step(wav_def, flux_def, where='mid', color='black')
ax.step(wav_def, err_def, where='mid', color='cyan')

ax.set_xlabel('Wavelength (Å)')
ax.set_ylabel(r'Flux (erg/s/cm${}^2$/Å)')

ax.set_title('J011035 (default, extrsize=7)')
```

```python
fig, ax = plt.subplots(1, figsize=(15,4))
ax.step(wav_def, flux_def, where='mid', color='black')
ax.step(wav_def, err_def, where='mid', color='cyan')

ax.set_xlabel('Wavelength (Å)')
ax.set_ylabel(r'Flux (erg/s/cm${}^2$/Å)')

ax.set_title('J011035 (default, extrsize=7)')

ax.set_xlim(2580,2610)
ax.set_ylim(-1.5e-15, 7e-15)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(15,5.), gridspec_kw={'height_ratios': [2.8, 1]}, sharex=True)

axes[0].step(wav_cust_1, flux_cust_1, where='mid', color='black', label='extrsize=3')
axes[0].step(wav_cust_1, err_cust_1, where='mid', color='cyan')

axes[0].step(wav_def, flux_def, where='mid', color='black', alpha=.5, linestyle='--', label='extrsize=7')
axes[0].step(wav_def, err_def, where='mid', color='cyan', alpha=.5, linestyle='--')

axes[0].set_xlim(2580,2610)
axes[0].set_ylim(-1.5e-15, 7e-15)

axes[0].set_title('J011035 (default, extrsize=3)')
axes[0].set_ylabel(r'Flux (erg/s/cm${}^2$/Å)')
axes[0].legend()

axes[1].step(wav_cust_1, (flux_cust_1-flux_def)/np.sqrt(err_cust_1**2+err_def**2), where='mid', 
             color='black')

axes[1].set_ylabel(r'$\Delta f / \sigma_f$')
axes[1].set_xlabel('Wavelength (Å)')
```
