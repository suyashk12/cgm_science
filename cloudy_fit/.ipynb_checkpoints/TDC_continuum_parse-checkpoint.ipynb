## TDC

Read in temperatures

```python
T_grid = np.loadtxt(rootdir+'CI/TDC_isochoric/log_metals=0/T0=1e8K/TDC_isochoric_frac.tim')[:,4]
```

Read emission and two-photon continuum

```python
with open(rootdir+'CI/TDC_isochoric/log_metals=0/T0=1e8K/TDC_isochoric_frac.conem') as file:
    L_em = [line.rstrip() for line in file]
    
with open(rootdir+'CI/TDC_isochoric/log_metals=0/T0=1e8K/TDC_isochoric_frac.contp') as file:
    L_tp = [line.rstrip() for line in file]
```

```python
wav_grid = []
nu_f_nu_total_grid = []
nu_f_nu_tp_grid = []

for i in range(1, len(L_em)):
    if '###########################' not in L_em[i]:
        L_em_split = L_em[i].split('\t')
        wav_grid.append(float(L_em_split[0]))
        nu_f_nu_total_grid.append(float(L_em_split[3]))
        
for i in range(1, len(L_tp)):
    if '###########################' not in L_tp[i]:
        L_tp_split = L_tp[i].split('\t')
        nu_f_nu_tp_grid.append(float(L_tp_split[2]))
```

Reshape

```python
n_steps_T = len(T_grid)
n_steps_wav = (len(L_em[1:])-n_steps_T)//n_steps_T
```

```python
wav_grid = np.array(wav_grid)
wav_grid = wav_grid.reshape((n_steps_T,n_steps_wav))

nu_f_nu_total_grid = np.array(nu_f_nu_total_grid)
nu_f_nu_total_grid = nu_f_nu_total_grid.reshape((n_steps_T,n_steps_wav))

nu_f_nu_tp_grid = np.array(nu_f_nu_tp_grid)
nu_f_nu_tp_grid = nu_f_nu_tp_grid.reshape((n_steps_T,n_steps_wav))
```

Convert to standard units

```python
f_lam_total_grid = nu_f_nu_total_grid/wav_grid 
f_lam_tp_grid = nu_f_nu_tp_grid/wav_grid
```

Make a plot to test

```python
logT_grid = np.log10(T_grid)
```

```python
fig, ax = plt.subplots(1, figsize=(4,4))

ax.step(wav_grid[0,:], np.log10(f_lam_total_grid[0,:]), where='mid', color='red', 
        label=r'$\log(T/\mathrm{K})=$'+' {}'.format(np.round(logT_grid[0],1)))
ax.step(wav_grid[100,:], np.log10(f_lam_total_grid[100,:]), where='mid', color='green', 
        label=r'$\log(T/\mathrm{K})=$'+' {}'.format(np.round(logT_grid[100],1)))
ax.step(wav_grid[200,:], np.log10(f_lam_total_grid[200,:]), where='mid', color='blue', 
        label=r'$\log(T/\mathrm{K})=$'+' {}'.format(np.round(logT_grid[200],1)))

ax.legend()

ax.axvline(1215.67, linestyle=':')
ax.set_xlim(1200,1230)
ax.set_ylim(-25.1, -24.7)
ax.set_ylim(-29, -21)
ax.set_xlabel(r'$\lambda$ (Å)')
ax.set_ylabel(r'$\log(f_\lambda/\mathrm{erg \ \mathrm{s}^{-1} \ \mathrm{cm}^{-2} \ \mathrm{Å}^{-1}})$')

ax.set_title(r'Ly$\alpha$ evolution for time-dependent cooling', fontsize=15)
#ax.set_ylim(-1e-30, 1e30)
#ax.axvline(1215)
```

```python
fig, ax = plt.subplots(1, figsize=(4,4))

ax.step(np.log10(wav_grid[0,:]), np.log10(f_lam_tp_grid[0,:]), where='mid', color='red', 
        label=r'$\log(T/\mathrm{K})=$'+' {}'.format(np.round(logT_grid[0],1)))
ax.step(np.log10(wav_grid[100,:]), np.log10(f_lam_tp_grid[100,:]), where='mid', color='green', 
        label=r'$\log(T/\mathrm{K})=$'+' {}'.format(np.round(logT_grid[100],1)))
ax.step(np.log10(wav_grid[200,:]), np.log10(f_lam_tp_grid[200,:]), where='mid', color='blue', 
        label=r'$\log(T/\mathrm{K})=$'+' {}'.format(np.round(logT_grid[200],1)))

ax.set_xlabel(r'$\log(\lambda/\mathrm{Å})$')
ax.set_ylabel(r'$\log(f_\lambda/\mathrm{erg \ \mathrm{s}^{-1} \ \mathrm{cm}^{-2} \ \mathrm{Å}^{-1}})$')

ax.legend()

ax.set_ylim(-33, -23)
ax.set_xlim(0,5)

ax.set_title(r'2-photon evolution for time-dependent cooling', fontsize=15)
#ax.set_xlim(900,1400)
#ax.set_ylim(-25.5, -24.6)
#ax.set_ylim(-1e-30, 1e30)
#ax.axvline(1215)
```

## CIE

Read in temperatures

```python
logT_grid_CIE = np.float_(np.loadtxt(rootdir+'CI/CIE/CIE_frac.grd', dtype=str)[:,6])
```

```python
with open(rootdir+'CI/CIE/CIE_frac.conem') as file:
    L_em = [line.rstrip() for line in file]
    
with open(rootdir+'CI/CIE/CIE_frac.contp') as file:
    L_tp = [line.rstrip() for line in file]
```

```python
wav_grid_CIE = []
nu_f_nu_total_grid_CIE = []
nu_f_nu_tp_grid_CIE = []

for i in range(1, len(L_em)):
    if '###########################' not in L_em[i]:
        L_em_split = L_em[i].split('\t')
        wav_grid_CIE.append(float(L_em_split[0]))
        nu_f_nu_total_grid_CIE.append(float(L_em_split[3]))
        
for i in range(1, len(L_tp)):
    if '###########################' not in L_tp[i]:
        L_tp_split = L_tp[i].split('\t')
        nu_f_nu_tp_grid_CIE.append(float(L_tp_split[2]))
```

Reshape

```python
n_steps_T = len(logT_grid_CIE)
n_steps_wav = (len(L_em[1:])-2*n_steps_T)//(2*n_steps_T)
```

```python
wav_grid_CIE_full = np.array(wav_grid_CIE)
wav_grid_CIE_full = wav_grid_CIE_full.reshape((2*n_steps_T,n_steps_wav))

nu_f_nu_total_grid_CIE_full = np.array(nu_f_nu_total_grid_CIE)
nu_f_nu_total_grid_CIE_full = nu_f_nu_total_grid_CIE_full.reshape((2*n_steps_T,n_steps_wav))

nu_f_nu_tp_grid_CIE_full = np.array(nu_f_nu_tp_grid_CIE)
nu_f_nu_tp_grid_CIE_full = nu_f_nu_tp_grid_CIE_full.reshape((2*n_steps_T,n_steps_wav))
```

```python
wav_grid_CIE = np.zeros((n_steps_T, n_steps_wav))
nu_f_nu_total_grid_CIE = np.zeros((n_steps_T, n_steps_wav))
nu_f_nu_tp_grid_CIE = np.zeros((n_steps_T, n_steps_wav))

for i in range(n_steps_T):
    wav_grid_CIE[i,:] = wav_grid_CIE_full[2*i+1,:]
    nu_f_nu_total_grid_CIE[i,:] = nu_f_nu_total_grid_CIE_full[2*i+1,:]
    nu_f_nu_tp_grid_CIE[i,:] = nu_f_nu_tp_grid_CIE_full[2*i+1,:]
```

Convert to standard units

```python
f_lam_total_grid_CIE = nu_f_nu_total_grid_CIE/wav_grid_CIE
f_lam_tp_grid_CIE = nu_f_nu_tp_grid_CIE/wav_grid_CIE
```

Make a plot to test

```python
idx1 = np.argmin(np.abs(logT_grid-8.0))
idx1_CIE = np.argmin(np.abs(logT_grid_CIE-8.0))

idx2 = np.argmin(np.abs(logT_grid-6.4))
idx2_CIE = np.argmin(np.abs(logT_grid_CIE-6.4))

idx3 = np.argmin(np.abs(logT_grid-4.6))
idx3_CIE = np.argmin(np.abs(logT_grid_CIE-4.6))
```

# TDC vs. CIE

```python
logT_grid[idx1]-logT_grid_CIE[idx1_CIE], logT_grid[idx2]-logT_grid_CIE[idx2_CIE], logT_grid[idx3]-logT_grid_CIE[idx3_CIE]
```

```python
fig, ax = plt.subplots(1, figsize=(5,5))

ax.step(wav_grid[idx1,:], np.log10(f_lam_total_grid[idx1,:]), where='mid', color='red', 
        label=r'$\log(T/\mathrm{K})=8.0$ (TDC)')
ax.step(wav_grid[idx2,:], np.log10(f_lam_total_grid[idx2,:]), where='mid', color='green', 
        label=r'$\log(T/\mathrm{K})=6.4$')
ax.step(wav_grid[idx3,:], np.log10(f_lam_total_grid[idx3,:]), where='mid', color='blue', 
        label=r'$\log(T/\mathrm{K})=4.6$')

ax.step(wav_grid_CIE[idx1_CIE,:], np.log10(f_lam_total_grid_CIE[idx1_CIE,:]), linestyle='--', where='mid', color='red', 
        label=r'$\log(T/\mathrm{K})=8.0$ (CIE)')
ax.step(wav_grid_CIE[idx2_CIE,:], np.log10(f_lam_total_grid_CIE[idx2_CIE,:]), linestyle='--', where='mid', color='green')
ax.step(wav_grid_CIE[idx3_CIE,:], np.log10(f_lam_total_grid_CIE[idx3_CIE,:]), linestyle='--', where='mid', color='blue')

ax.legend()

ax.axvline(1215.67, linestyle=':')
ax.set_xlim(1200,1230)
ax.set_ylim(-25.1, -24.7)
ax.set_ylim(-29, -21)
ax.set_xlabel(r'$\lambda$ (Å)')
ax.set_ylabel(r'$\log(f_\lambda/\mathrm{erg \ \mathrm{s}^{-1} \ \mathrm{cm}^{-2} \ \mathrm{Å}^{-1}})$')

ax.set_title(r'Ly$\alpha$ evolution (TDC vs. CIE)', fontsize=15)
```

```python
fig, ax = plt.subplots(1, figsize=(5,5))

ax.step(wav_grid[idx1,:], f_lam_total_grid[idx1,:]/f_lam_total_grid_CIE[idx1_CIE,:], 
        where='mid', color='red', label=r'$\log(T/\mathrm{K})=8.0$')
ax.step(wav_grid[idx2,:], f_lam_total_grid[idx2,:]/f_lam_total_grid_CIE[idx2_CIE,:], 
        where='mid', color='green', label=r'$\log(T/\mathrm{K})=6.4$')
ax.step(wav_grid[idx3,:], f_lam_total_grid[idx3,:]/f_lam_total_grid_CIE[idx3_CIE,:], 
        where='mid', color='blue', label=r'$\log(T/\mathrm{K})=4.6$')

ax.legend()

ax.axhline(1, linestyle='--')

ax.axvline(1215.67, linestyle=':')
ax.set_xlim(1200,1230)
ax.set_ylim(0, 10)
ax.set_xlabel(r'$\lambda$ (Å)')
ax.set_ylabel(r'$f_{\lambda, \mathrm{TDC}}/f_{\lambda, \mathrm{CIE}}$')

ax.set_title(r'Ly$\alpha$ evolution (TDC vs. CIE)', fontsize=15)
```

```python
fig, ax = plt.subplots(1, figsize=(5,5))

ax.step(np.log10(wav_grid[idx1,:]), np.log10(f_lam_tp_grid[idx1,:]), where='mid', color='red', 
        label=r'$\log(T/\mathrm{K})=8.0$ (TDC)')
ax.step(np.log10(wav_grid[idx2,:]), np.log10(f_lam_tp_grid[idx2,:]), where='mid', color='green', 
        label=r'$\log(T/\mathrm{K})=6.4$')
ax.step(np.log10(wav_grid[idx3,:]), np.log10(f_lam_tp_grid[idx3,:]), where='mid', color='blue', 
        label=r'$\log(T/\mathrm{K})=4.6$')

ax.step(np.log10(wav_grid_CIE[idx1_CIE,:]), np.log10(f_lam_tp_grid_CIE[idx1_CIE,:]), linestyle='--', where='mid', color='red', 
        label=r'$\log(T/\mathrm{K})=8.0$ (CIE)')
ax.step(np.log10(wav_grid_CIE[idx2_CIE,:]), np.log10(f_lam_tp_grid_CIE[idx2_CIE,:]), linestyle='--', where='mid', color='green')
ax.step(np.log10(wav_grid_CIE[idx3_CIE,:]), np.log10(f_lam_tp_grid_CIE[idx3_CIE,:]), linestyle='--', where='mid', color='blue')

ax.set_xlabel(r'$\log(\lambda/\mathrm{Å})$')
ax.set_ylabel(r'$\log(f_\lambda/\mathrm{erg \ \mathrm{s}^{-1} \ \mathrm{cm}^{-2} \ \mathrm{Å}^{-1}})$')

ax.legend()

ax.set_ylim(-33, -23)
ax.set_xlim(0,5)

ax.set_title(r'2-photon evolution (TDC vs. CIE)', fontsize=15)
#ax.set_xlim(900,1400)
#ax.set_ylim(-25.5, -24.6)
#ax.set_ylim(-1e-30, 1e30)
#ax.axvline(1215)
```

```python
fig, ax = plt.subplots(1, figsize=(5,5))

ax.step(np.log10(wav_grid[idx1,:]), f_lam_tp_grid[idx1,:]/f_lam_tp_grid_CIE[idx1_CIE,:], 
        where='mid', color='red', label=r'$\log(T/\mathrm{K})=8.0$')
ax.step(np.log10(wav_grid[idx2,:]), f_lam_tp_grid[idx2,:]/f_lam_tp_grid_CIE[idx2_CIE,:], 
        where='mid', color='green', label=r'$\log(T/\mathrm{K})=6.4$')
ax.step(np.log10(wav_grid[idx3,:]), f_lam_tp_grid[idx3,:]/f_lam_tp_grid_CIE[idx3_CIE,:], 
        where='mid', color='blue', label=r'$\log(T/\mathrm{K})=4.6$')

ax.set_xlabel(r'$\log(\lambda/\mathrm{Å})$')
ax.set_ylabel(r'$f_{\lambda, \mathrm{TDC}}/f_{\lambda, \mathrm{CIE}}$')

ax.legend()

ax.set_xlim(0,5)
ax.set_ylim(0, 10)


ax.set_title(r'2-photon evolution (TDC vs. CIE)', fontsize=15)
#ax.set_xlim(900,1400)
#ax.set_ylim(-25.5, -24.6)
#ax.set_ylim(-1e-30, 1e30)
#ax.axvline(1215)
```
