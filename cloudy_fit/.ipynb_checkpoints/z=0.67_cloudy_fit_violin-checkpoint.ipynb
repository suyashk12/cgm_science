Define the log likehood function

```python
def log_prior(params):
    
    # Grid parameters being varied
    log_hdens, log_metals = params
    
    # If the sampled density is within the CLOUDY limits
    # Avoid edges?
    if log_hdens_min<log_hdens<log_hdens_max and log_metals_min<log_metals<log_metals_max:
        return 0.0
    return -np.inf
```

```python
def log_likelihood(params, logN_HI, logN_pdf_interps, species_logN_interp):
    
    log_hdens, log_metals = params
    
    # First get the estimated column densities
    ll = 0
    
    ions = list(logN_pdf_interps.keys())
    
    for i in range(len(ions)):
        
        ion = ions[i]
        s = ion_species_dict[ion]
        
        # Get interpolated column density from CLOUDY grid
        logN_s = species_logN_interp[s]([logN_HI, log_hdens, log_metals])[0]
        
        # Get posterior probability from interpolated VP fit posterior
        # Take natural log, not log10
        ll += np.log(max(logN_pdf_interps[ion](10**logN_s),1e-25))
    return ll
```

```python
def log_probability(params, logN_HI, logN_pdf_interps, species_logN_interp):
    
    lp = log_prior(params)
    
    if not np.isfinite(lp):
        return -np.inf
    
    return lp + log_likelihood(params, logN_HI, logN_pdf_interps, species_logN_interp)
```
