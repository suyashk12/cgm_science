```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python
x_arr = np.linspace(0, 100, 100)
dof_arr = np.linspace(40, 60, 10)
```

```python
pdf_arr = np.array([chi2.pdf(x_arr, dof) for dof in dof_arr])
```

```python
pdf_arr.shape
```

```python
fig, ax = plt.subplots(1, figsize=(5,4))

for i in range(len(pdf_arr)):
    ax.plot(x_arr, pdf_arr[i,:])
    print(chi2.ppf(0.84, dof_arr[i])-chi2.ppf(0.5, dof_arr[i]))

ax.set_xlabel(r'$\chi^2$')
ax.set_ylabel(r'PDF')
```
