```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
```

```python
fig, ax = plt.subplots(1, figsize=(15, 7))

ax.plot(x, y)
ax.plot(x, np.gradient(y,x))
ax.plot(x, np.gradient(np.gradient(y,x),x))


ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
def list_local_minima(x, f):
    
    f_grad = np.gradient(f,x)
    
    minima_loc_list = []
    minima_list = []
    minima_dgrad_list = []
    
    for i in range(len(f)-1):
        
        if(f_grad[i]<0 and f_grad[i+1]>0):
            
            x1 = x[i]
            x2 = x[i+1]
            y1 = f_grad[i]
            y2 = f_grad[i+1]
            
            m = (y2-y1)/(x2-x1)
            b = y1-m*x1
            
            x0 = -b/m
            
            f_dgrad_x0 = (y2-y1)/(x2-x1)
            
            f_x0 = f[i] + m*(x0-x1)**2 + b*(x0-x1)
            
            minima_loc_list.append(x0)
            minima_list.append(f_x0)
            minima_dgrad_list.append(f_dgrad_x0)
```
